```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot

# Load your dataset
file_path = 'dataset.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Specify the column with time series data (e.g., 'pre_usage' or 'post_usage')
unit_columns = [col for col in data.columns if 'unit' in col]# Replace with the actual column name for prepaid usage

# unit_columns = ["pre_22_june_unit"]

# Create lag plots for each unit column
for column in unit_columns:
    plt.figure(figsize=(6, 6))
    lag_plot(data[column])
    plt.title(f'Lag Plot for {column}')
    plt.xlabel(f'{column} (t)')
    plt.ylabel(f'{column} (t+1)')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"Lag Plot for {column}.png", dpi=300)
    plt.show()
```
